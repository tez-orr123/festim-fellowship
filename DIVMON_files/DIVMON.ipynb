```python
# Exhaust monoblock (DIVMON)

# 5mm W, 1mm Cu, 2mm CuCrZr 
# 1.2 MW/m2 steady state heat flux
# 1.3-28.8 MW/m3 neutronic heating for W
# 0.3-8.1 MW/m3 neutronic heating for Cu and CuCrZr
# 473-480 K coolant boundary
# 0.1-2 dpa/fpy W neutronic damage
# 0.5-13.9 dpa.fpy CuCrZr neutronic damage
```

```python
import festim as F


my_model = F.Simulation()

import numpy as np

my_model.mesh = F.MeshFromVertices(np.linspace(0, 8e-3, num=100000))

tungsten = F.Material(
    id=1,
    D_0=1.5e-07,  # m2/s
    E_D=0.265,  # eV
    S_0=2.7e24,
    E_S=1.14,
    borders=[0,5e-3],
    thermal_cond=173,  # W/mK
)

copper = F.Material(
    id=2,
    D_0=6.6e-7,
    E_D=0.387,
    S_0=3.14e24,
    E_S=0.572,
    borders=[5e-3, 6e-3],
    thermal_cond=350,
)

cucrzr = F.Material(
	id=3, 
	D_0=4.8e-7, 
	E_D=0.42, 
	S_0=4.27e23, 
	E_S=0.39,
	borders=[6e-3, 8e-3],
	thermal_cond=320
)

my_model.materials = [tungsten, copper, cucrzr]
```

```python
w_density = 6.3e28  # atom/m3  
copper_density = 9.2e28
cucrzr_density = 9.2e28  

damage_percent = 0.01 # 0.1-2 dpa/fpy W neutronic damage

trap_W_1 = F.Trap( 
	k_0 = [1.5e-7/((1.1e-10)**2 * 6 * w_density),6.6e-7 / ((3.6e-10)**2 * 1 * copper_density),4.8e-7 / ((3.6e-10)**2 * 1 * cucrzr_density)],
	E_k= [0.265,0.387,0.418],
	p_0=[1.2397e13,5.0926e12,7.3472e12],
	E_p=[0.83,0.5,0.53], 
	density = [w_density*0.00118,copper_density*0.00005,3.7e24],
	materials = [tungsten,copper,cucrzr]
)
            
trap_W_2 = F.Trap(
	k_0 = 1.5e-7/((1.1e-10)**2 * 6 * w_density),
	E_k= 0.265, 
	p_0=1.2397e13,
	E_p=0.97,
	density = w_density*0.000722,
	materials = [tungsten]
)

trap_W_3 = F.Trap(
	k_0=1.5e-7/(1.1e-10**2 * 6 * w_density),
	E_k=0.265,
	p_0=1e13,
	E_p=1.51, # change to 2.05 when necessary
	density = damage_percent*w_density,
	materials = tungsten
) 
my_model.traps = [trap_W_1, trap_W_2, trap_W_3]
```

```python
my_model.T = F.HeatTransferProblem(transient=False)

my_model.boundary_conditions = [
    F.ImplantationDirichlet(surfaces=1, phi=0.23e24, R_p=1.1e-9, D_0=1.5e-07, E_D=0.265), # mean depth may be too low, maybe order of magnitude higher
    F.DirichletBC(surfaces=2, value=0, field=0),
    #F.FluxBC(surfaces=1, value=10e6, field="T"),
    F.DirichletBC(surfaces=1, value=873, field="T"),
    F.DirichletBC(surfaces=2, value=570, field="T"), # Set coolant temp to 573K as 473K messes it up for some reason
]
# boundary conditions on the left (plasma facing) side can be defined as having fixed temperature
# of 1273K using DirichletBC or a heat flux of 12e6 using FluxBC. The value of heat flux needs to
# be order of magnitude higher than what Dinusha initially said (1.2e6) to see saturation in 500 days.
# Localised heating may be much higher than calculated so the design case is 10 MW/m2. 
my_model.sources = [ 
		F.Source(value=28.8e6, volume=1, field='T'), # Neutronic heating W: 1.3-28.8 MW/m3
		F.Source(value=8.1e6, volume=[2, 3], field='T'), # Neutronic heating Cu & CuCrZr: 0.3-8.1 MW/m3
]
		
# Stepsize
my_model.dt = F.Stepsize(
    initial_value=0.5,
    stepsize_change_ratio=1.1
)


# Settings
my_model.settings = F.Settings(
    absolute_tolerance=1e14, relative_tolerance=1e-04, final_time=3.2e7, # relative tolerance has been greatly decreased
    # if you get the float error, change the flux value (phi) in ImplantationDirichlet to a lower magnitude instead of
    # changing tolerance values.
    chemical_pot=True, traps_element_type="DG"
)

retention_1 = F.TotalVolume("retention", volume = 1)
retention_2 = F.TotalVolume("retention", volume = 2)
retention_3 = F.TotalVolume("retention", volume = 3)
solute_1 = F.TotalVolume("solute", volume = 1)
solute_2 = F.TotalVolume("solute", volume = 2)
solute_3 = F.TotalVolume("solute", volume = 3) # Changed this to include mobile concentration to get full inventory of H

derived_quantities = F.DerivedQuantities(
    [retention_1, retention_2, retention_3, solute_1, solute_2, solute_3]
)

results_folder = "PFC_retention_profiles/DIVMON"

txt_export = F.TXTExport(
    field='T',
    filename=results_folder + '/T.txt',
    times=[3.2e7],
    )

my_model.exports = [F.DerivedQuantities([retention_1, retention_2, retention_3, solute_1, solute_2, solute_3], 
                                        filename = results_folder + "/derived_qs.csv"), txt_export]

my_model.initialise()
my_model.run()
```

```python
import matplotlib.pyplot as plt

retention_full = (
    np.array(retention_1.data)
    + np.array(retention_2.data)
    + np.array(retention_3.data)
)

solute_full = (
    + np.array(solute_1.data)
    + np.array(solute_2.data)
    + np.array(solute_3.data) # Changed this to include mobile concentration to get full inventory of H
)


retention_a = retention_full / 6.022e23
retention_b = retention_a / 2
retention_c = retention_b * 3
PFC_area = 138.68 # m^2
retention_final = retention_c * PFC_area

plt.plot(retention_1.t, retention_final, label="DIVMON", color="tab:green")
plt.legend()
plt.ylabel("Trapped Inventory (g)")
plt.xlabel("Time (s)")
plt.show()


solute_a = solute_full / 6.022e23
solute_b = solute_a / 2
solute_c = solute_b * 3
PFC_area = 138.68 # m^2
solute_final = solute_c * PFC_area

plt.plot(solute_1.t, solute_final, label="DIVMON", color="tab:green")
plt.legend()
plt.ylabel("Mobile Inventory (g)")
plt.xlabel("Time (s)")
plt.show()

```
